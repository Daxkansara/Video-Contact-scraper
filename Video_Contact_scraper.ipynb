```python
import pytesseract
from PIL import Image
import os

texts = []
for file in sorted(os.listdir("frames")):
    if file.endswith(".jpg"):
        text = pytesseract.image_to_string(Image.open("frames/"+file))
        if text.strip():
            texts.append(text.strip())

with open("contacts.txt", "w", encoding="utf-8") as f:
    f.write("\n".join(texts))
```

```python
import re
import pandas as pd

cleaned = []
pattern = re.compile(r'([A-Za-z ]+)\s+(\+?\d[\d\s\-]{7,})')

for line in texts:
    matches = pattern.findall(line)
    for m in matches:
        name = m[0].strip()
        phone = m[1].strip().replace(" ", "")
        cleaned.append([name, phone])

df = pd.DataFrame(cleaned, columns=["Name", "Phone"])
df.to_excel("contacts_cleaned.xlsx", index=False)
from google.colab import files
files.download("contacts_cleaned.xlsx")
```
